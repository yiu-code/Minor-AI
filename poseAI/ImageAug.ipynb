```python
import os
import pandas as pd
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

%matplotlib inline
```

```python
class ImageAugmentation(object):   
    
    def import_and_mirror_images(self, dir_label):
        for obj in dir_label:
            for filename in os.scandir(obj['dir']):
                if (filename.path.endswith(".jpg") or filename.path.endswith(".png")) and filename.is_file():
                    print(filename)
                    img = cv2.imread(filename.path)
                    height, width = img.shape[:2]
#                     print(Path(filename).stem)

                    complete_img = cv2.flip(img, 1) # Flips image horizontally
                    cv2.imwrite(obj['dir'] + Path(filename).stem +"_mirrored.jpg", complete_img) # Filenaam nog aan toevoegen
```
