```python
import os

from PIL import Image
import requests
import urllib.request as urllib2
import shutil
from io import BytesIO
import pandas as pd

df = pd.read_csv("Tree_Pose_or_Vrksasana_.txt", error_bad_lines=False, names=["fileName", "urlPath"])


def loop_through_df(df):
    name = ""
    url = ""
    
    for index in range(len(df.iloc[1:])):
        name = df.iloc[index,0]
        url = df.iloc[index,1]
        if(os.path.exists("treePose/" + os.path.basename(name))):
            continue
        try:
            response = requests.get(url)
            if response.status_code != 200:
                print(response.status_code)
                print(name)
                continue
            img = Image.open(BytesIO(response.content))
            print(os.path.basename(name))
            response = img.save("treePose/" + os.path.basename(name))
        except:
            continue


loop_through_df(df)
```
